## Configure Paths and Controls

```python
MODEL_PATH = '/content/drive/MyDrive/models/your_model.gguf'
OUT_DIR = '/content/drive/MyDrive/mafia_runs/run_001'

N_GAMES = 2000
PARALLEL_GAMES = 8  # try 4 first on free tier, then increase
PLAYERS = 9
MAFIA_COUNT = 2
MAX_ROUNDS = 10
TALKS_PER_DAY = 1

NIGHT_MODE = 'model'  # 'random' or 'model'
RESUME = True
COMPRESS_GAMES = True

TEMPERATURE = 0.95
TOP_P = 0.90
TOP_K = 50
MIN_P = 0.05
REPEAT_PENALTY = 1.1
BASE_SEED = 20260305

N_CTX = 2048
N_THREADS = 2
N_GPU_LAYERS = -1
N_BATCH = 512
```

```python
import os
cmd = [
    'python', 'mafia_self_play.py',
    '--model_path', MODEL_PATH,
    '--out_dir', OUT_DIR,
    '--n_games', str(N_GAMES),
    '--parallel_games', str(PARALLEL_GAMES),
    '--players', str(PLAYERS),
    '--mafia_count', str(MAFIA_COUNT),
    '--max_rounds', str(MAX_ROUNDS),
    '--talks_per_day', str(TALKS_PER_DAY),
    '--night_mode', NIGHT_MODE,
    '--temperature', str(TEMPERATURE),
    '--top_p', str(TOP_P),
    '--top_k', str(TOP_K),
    '--min_p', str(MIN_P),
    '--repeat_penalty', str(REPEAT_PENALTY),
    '--base_seed', str(BASE_SEED),
    '--n_ctx', str(N_CTX),
    '--n_threads', str(N_THREADS),
    '--n_gpu_layers', str(N_GPU_LAYERS),
    '--n_batch', str(N_BATCH),
]
if RESUME:
    cmd.append('--resume')
if COMPRESS_GAMES:
    cmd.append('--compress_games')
print(' '.join(cmd))
ret = os.system(' '.join(cmd))
print('exit code:', ret)
```
